```python
# Importing Libraries
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sb
import lasio
```

```python
# read log file
df = pd.read_csv("Dutch_F3_Logs\Exported.csv")
#1. Investigate the component of the data file (number of columns , numbers of observations, Null values, normal statistics)
display(df)
```

```python
df.info()
```

```python
# null values
df.isnull().sum()
```

```python
# statistics
df.describe()
```

```python
#2. plot the null values as bars
df.isnull().sum().plot(kind = 'bar', title = 'Missing Values Comparison', figsize=(10,5)); # pipeline() # %>%
```

```python
#3. fill in null values using three different methods
#4. produce statistics of the different methods and plot them
#a. fill by mean value
df_edited=df.fillna(df.mean())
df_edited.describe()
```

```python
#3. fill in null values using three different methods
#4. produce statistics of the different methods and plot them
#b. fill by interpolation
df_edited_2=df.interpolate(method='linear', limit_direction='backward')
df_edited_2.describe()
```

```python
#3. fill in null values using three different methods
#4. produce statistics of the different methods and plot them
#c. fill by interpolation (polynomial)
df_edited_3=df.interpolate(method='polynomial', order=2, limit_direction='backward')
df_edited_3.describe()
```

```python
#plot the output
plt.figure(figsize=(15,5))
plt.hist(df_edited.RHOB, color='g', bins = 100)
plt.title('RHOB_#1 Histogram ')
plt.xlabel('RHOB Values ')
plt.ylabel('Frequency')
plt.show();

plt.figure(figsize=(15,5))
plt.hist(df_edited_2.RHOB, color='r', bins = 100)
plt.title('RHOB_#2 Histogram ')
plt.xlabel('RHOB Values ')
plt.ylabel('Frequency')
plt.show();

plt.figure(figsize=(15,5))
plt.hist(df_edited_3.RHOB, color='b', bins = 100)
plt.title('RHOB_#3 Histogram ')
plt.xlabel('RHOB Values ')
plt.ylabel('Frequency')
plt.show();
```

```python
#Plot the three Facies scenarios vs Depth
plt.figure(figsize=(15,15))

plt.subplot(1, 3, 1)
plt.plot(df_edited.RHOB, -1*df_edited.DEPTH, label = 'GR', c = 'g')
plt.xlabel('RHOB_#1')
plt.ylabel('DEPTH')

plt.subplot(1, 3, 2)
plt.plot(df_edited_2.RHOB, -1*df_edited_2.DEPTH, label = 'Facies1', c = 'b')
plt.xlabel('RHOB_#2')
plt.ylabel('DEPTH')

plt.subplot(1, 3, 3)
plt.plot(df_edited_3.RHOB, -1*df_edited_3.DEPTH, label = 'Facies2', c = 'r')
plt.xlabel('RHOB_#3')
plt.ylabel('DEPTH');
```
