```python
# Importing Libraries
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sb
import lasio
```

```python
# read log file
df = pd.read_csv("Dutch_F3_Logs\Exported.csv")
#1. Investigate the component of the data file (number of columns , numbers of observations, Null values, normal statistics)
display(df)
```

```python
# null values
df.isnull().sum()
```

```python
# statistics
df.describe()
```

```python
#2. plot the null values as bars
df.isnull().sum().plot(kind = 'bar', title = 'Missing Values Comparison', figsize=(10,5)); # pipeline() # %>%
```

```python
#3. fill in null values using three different methods
#a. fill by mean value
df_edited=df.fillna(df.mean())
df_edited.describe()
```

```python
#3. fill in null values using three different methods
#b. fill by interpolation
df_edited_2=df.interpolate(method='linear', limit_direction='backward')
df_edited_2.describe()
```

```python
#3. fill in null values using three different methods
#c. fill by interpolation (polynomial)
df_edited_3=df.interpolate(method='polynomial', order=2, limit_direction='backward')
df_edited_3.describe()
```

```python
#4. Create three facies Log for with 3 different scenarios
#a. scenario1
def SH(A):
    return 'shale' if A>15 else 'sand'
df_edited['Facies1']=df_edited['GR'].apply(SH)
df_edited_dummy=pd.get_dummies(df_edited['Facies1'])

#b. scenario2
def SH2(A):
    return 'shale' if A>20 else 'sand'
df_edited['Facies2']=df_edited['GR'].apply(SH2)
df_edited_dummy2=pd.get_dummies(df_edited['Facies2'])

#c. scenario3
def SH3(A):
    return 'shale' if A>25 else 'sand'
df_edited['Facies3']=df_edited['GR'].apply(SH3)
df_edited_dummy3=pd.get_dummies(df_edited['Facies3'])

# merge facies
df_edited_facies=pd.concat([df_edited, df_edited_dummy, df_edited_dummy2, df_edited_dummy3], axis=1)
df_edited_facies
```

```python
#5. Plot the three Facies scenarios vs Depth
plt.figure(figsize=(15,15))

plt.subplot(1, 4, 1)
plt.plot(df_edited_facies.GR, -1*df_edited_facies.DEPTH, label = 'GR', c = 'g')
plt.xlabel('GR')
plt.ylabel('DEPTH')

plt.subplot(1, 4, 2)
plt.plot(df_edited_facies.Facies1, -1*df_edited_facies.DEPTH, label = 'Facies1', c = 'b')
plt.xlabel('Facies1')
plt.ylabel('DEPTH')
plt.fill_between(df_edited_facies.Facies1, -1*df_edited_facies.DEPTH, 1, alpha=0.5)

plt.subplot(1, 4, 3)
plt.plot(df_edited_facies.Facies2, -1*df_edited_facies.DEPTH, label = 'Facies2', c = 'r')
plt.xlabel('Facies2')
plt.ylabel('DEPTH')
plt.fill_between(df_edited_facies.Facies2, -1*df_edited_facies.DEPTH, 1, alpha=0.5)

plt.subplot(1, 4, 4)
plt.plot(df_edited_facies.Facies3, -1*df_edited_facies.DEPTH, label = 'Facies3', c = 'black')
plt.xlabel('Facies3')
plt.ylabel('DEPTH')
plt.fill_between(df_edited_facies.Facies3, -1*df_edited_facies.DEPTH, 1, alpha=0.5);
```
