```python
# Importing Libraries
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sb
import lasio
```

```python
# read log file
df = pd.read_csv("well1513.csv")
#1. Investigate the component of the data file (number of columns , numbers of observations, Null values, normal statistics)
display(df)
```

```python
df.info()
```

```python
# null values
df.isnull().sum()
```

```python
# statistics
df.describe()
```

```python
# dataframe copy
df_copy=df
# drop the invalid columns
df_copy.drop(columns=['FORMATION', 'RSHA', 'SGR', 'BS', 'DTS', 'DCAL', 'RMIC', 'ROPA', 'RXO'], inplace=True)
# drop nan values
df_copy.dropna(inplace=True)
df_copy.isna().sum()
```

```python
# MD to be the index
df_copy.set_index('DEPTH_MD', inplace=True)
df_copy
```

```python
#2. Create subset including (DT, NPHI, GR,SP) logs
df_subset=df_copy[['DTC', 'NPHI', 'GR', 'SP']]
df_subset
```

```python
#3. Plot the hisogram of the subset
df_subset.hist(figsize=(15,10), bins=50);
```

```python
#3. Plot the hisogram of the subset
df_subset.plot(kind='hist',
        alpha=0.5,
        bins=50,
        title='Histogram Of the subset',
        rot=45,
        grid=True,
        figsize=(15,7),
        fontsize=12, 
        color=['g', 'r', 'orange', 'b']);
```

```python
#4. Create a function to normalize the well logs using formula (X-Xmin / Xmax-Xmin)
df_norm = (df_subset - df_subset.min()) / (df_subset.max() - df_subset.min())
df_norm
```

```python
#5. Plot histogram of the normalized well logs
df_norm.hist(figsize=(15,10), bins=50);
```

```python
#5. Plot histogram of the normalized well logs
df_norm.plot(kind='hist',
        alpha=0.5,
        bins=50,
        title='Histogram Of the normalized well logs',
        rot=45,
        grid=True,
        figsize=(15,7),
        fontsize=12, 
        color=['g', 'r', 'orange', 'b']);
```
